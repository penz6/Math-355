## Pandas library

Pandas is an external library written for the Python programming language for data manipulation and analysis. In particular, it offers data structures and operations for manipulating numerical tables and time series (soruce: Wiki).


There are two important data structures in Pandas, called **Series** and **DataFrame**.

- **Series**: Series is a one-dimensional labeled array capable of holding any Python data type (integers, strings, floating point numbers, etc.). The axis labels are collectively referred to as the index.

- **DataFrame**: DataFrame is a two-dimensional, size-mutable, potentially heterogeneous tabular data structure. It is a primary Pandas data structure.This data structure also contains labeled axes (rows and columns). Can be thought of as a dict-like container for Series objects.
