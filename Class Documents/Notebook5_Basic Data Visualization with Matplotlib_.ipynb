## Matplotlib

Matplotlib is a Python package for 2D and 3D-graphics. It provides a very quick way to visualize data from Python.


```python
import numpy as np

import matplotlib.pyplot as plt
```

### Simple Plots

```python
x = np.linspace(-np.pi, np.pi, num = 256) #Creating equispace points 
y = np.sin(x)

#Creating plotting window
plt.figure()
plt.plot(x,y)
plt.show()
```

### Customizing Figure Size

```python
plt.figure(figsize = (6,4))

plt.plot(x,y)

plt.show ()
```

### Multiple Plots 
It is often possible to construct multiple plots in the same figure

```python
x = np.linspace(-np.pi, np.pi, num = 256)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure()
plt.plot(x,y1)
plt.plot(x,y2)
plt.show( )
```

### Markers
It is sometimes relevant to show the data points.

[Python Marker](https://matplotlib.org/api/markers_api.html#module-matplotlib.markers)

```python
x = np.arange(0., 5., 0.2)
y1 = x
y2 = x**2
y3 = x**3

# red circle, blue squares and green triangles
plt.figure(figsize = (8,4))

plt.plot(x, y1, color = "r", marker = "o", markersize = 6)
plt.plot(x, y2, color = "b", marker = "s", markersize = 6)
plt.plot(x, y3, color=  "g", marker = "^", markersize = 8)
#plt.plot(x, y1, 'ro', x, y2, 'bs', x, y3, 'g^')
plt.show()
```

### Adding Labels,  Title, and Legend

```python
x = np.linspace(-np.pi, np.pi, num = 256) 
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize = (5,3))
plt.plot(x, y1, color = "r")
plt.plot(x, y2, color=  "b")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.legend(loc = "best", labels = ("sin x", "cos x"))

plt.title("Sine and Cosine graph")
plt.show()
```

### Custimizing Axis  and Grid

```python
plt.figure()
plt.plot(x, y1, color = "r")
plt.plot(x, y2, color=  "b")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
#plt.xlim(-2, 2)
#plt.ylim(-1.5, 1.5)
plt.xlim(np.min(x), np.max(x))
plt.ylim(np.min(y1),np.max(y1))

plt.legend(loc = "best", labels = ("sin x", "cos x"))

plt.title("Sine and Cosine graph")

plt.grid(True)

plt.show()
```

### Alternative Way of Customizing Axis

```python
plt.figure()
plt.plot(x, y1, color = "r")
plt.plot(x, y2, color=  "b")
plt.legend(loc = "best", labels = ("sin x", "cos x"))
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.axis([-2, 2, -1.5, 1.5]) #[x-axis-lower, x-axis-upper, y-axis-lower, y-axis-upper]

plt.title("Sine and Cosine graph")
plt.grid(True)
plt.show()
```

### Creating Subplots

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)


plt.figure(figsize = (5,4))

plt.subplot(211)
plt.plot(t1, f(t1), color = 'red', marker = "o")
plt.plot(t2, f(t2), color = 'blue')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), color = 'red', linestyle = '--')
plt.show()
```

```python
plt.figure(figsize = (10,4))
plt.subplot(121)
plt.plot(t1, f(t1), color = 'red', marker = "o")
plt.plot(t2, f(t2), color = 'blue')

plt.subplot(122)
plt.plot(t2, np.cos(2*np.pi*t2), color = 'red', linestyle = '--')
plt.show()
```

## Setting ticks

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
cos = np.cos(x)
sin = np.sin(x)



plt.figure(figsize=(5, 3), dpi = 80)

plt.plot(x, cos, color = "blue", linewidth = 2.5, linestyle = "-")
plt.plot(x, sin, color = "red",  linewidth = 2.5,  linestyle = "-")

plt.xlim(x.min() * 1.1, x.max() * 1.1)
plt.ylim(cos.min() * 1.1, cos.max() * 1.1)

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, 1])

plt.show()
```

```python
plt.figure(figsize=(7, 4), dpi=80)
plt.plot(x, cos, color="blue", linewidth=2.5, linestyle="-")
plt.plot(x, sin, color="red",  linewidth=2.5, linestyle="-")
plt.xlim(x.min() * 1.1, x.max() * 1.1)
plt.ylim(cos.min() * 1.1, cos.max() * 1.1)

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
          ['$-\pi$', '$-\pi/2$', '$0$', '$+\pi/2$', '$+\pi$'])

plt.yticks([-1, 0, 1], ['$-1$', '$0$', '$+1$'])

plt.title('$y = \sin x, y = \cos x$')
plt.show()
```

### Meshgrid and Contour Plots

```python
def g(x,y):
    return (x**2 + y**2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)


X, Y = np.meshgrid(x, y)

plt.figure(figsize = (6,5))

plt.contour(X, Y, g(X, Y), 9, colors='blue')
plt.show()
```

```python

plt.figure(figsize = (7,5))
plt.contour(X, Y, g(X, Y), 9, colors ='black')

plt.contourf(X, Y, g(X, Y), 9, alpha=.75, cmap = 'jet') # color map 
plt.show()
```

```python
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 -y ** 2)
n = 256

x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)


X, Y = np.meshgrid(x, y)

plt.figure(figsize = (8,5))
plt.contour(X, Y, f(X, Y), 9, colors='black')
plt.contourf(X, Y, f(X, Y), 9, alpha=.75, cmap='jet')
plt.show()
```

## Imshow

```python
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 - y ** 2)

n = 10
x = np.linspace(-3, 3, 4 * n)
y = np.linspace(-3, 3, 3 * n)
X, Y = np.meshgrid(x, y)

plt.imshow(f(X,Y))
plt.show()
```

##  3D Plots

```python
import warnings
warnings.filterwarnings('ignore')

from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter

#Create data.
x = np.arange(-5, 5, 0.25)
y = np.arange(-5, 5, 0.25)

X, Y = np.meshgrid(x, y)

R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)


fig = plt.figure(figsize = (15, 15))
                 
ax  = fig.gca(projection='3d')
                            
# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap = 'jet')

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink = 0.7, aspect = 5)
plt.show()
```
