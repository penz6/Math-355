## Reading Diamond Data

## Adding New Column to  the Dataframe

```python
diamonds_df['price_per_carat']  = diamonds_df['price']/diamonds_df['carat']

diamonds_df.head()
```

## Use of Conditional Statement While Creating New Column

```python
diamonds_df['price_per_carat_is_high'] = np.where(diamonds_df['price_per_carat']>1500, 1, 0)
```

# Creating Bar Plots


Use of Seaborn functions:

- **sns.barplot()** for making bar plot 
- **sns.catplot()** for making plot for categorical data

```python
ax = sns.barplot(x ="cut", y="price", data = diamonds_df, estimator= np.mean)
```

```python
ax = sns.barplot(x="cut", y="price", data=diamonds_df, estimator = np.mean, order=['Ideal','Good','Very Good','Fair','Premium'])
```

```python
ax = sns.catplot("cut", data = diamonds_df, aspect = 1.3,  kind="count", color="b")# categorical plot(bar plot)
```

```python
ax = sns.barplot(x="cut", y="price", data=diamonds_df, estimator=np.mean, order=['Ideal','Good','Very Good','Fair','Premium'])
```

### More Customization Options using Matpotlib functionaliy


```python
plt.figure(figsize = (10,5))
sns.barplot(x="cut", y="price", data=diamonds_df, estimator=np.mean, order=['Ideal','Good','Very Good','Fair','Premium'])
plt.title("Analysis of Diamond dataset")
plt.show()
```

## Creating Bar Plots Grouped by a Specific Feature

```python
plt.figure(figsize = (10, 5))
sns.barplot(x="cut", y="price", hue ='color', data= diamonds_df) # look at the third argument hue = 'color'
plt.show()
```

## Tweaking Plot Parameters of a Grouped Bar Plot

```python
#Modifying legends
plt.figure(figsize = (10, 5))
ax = sns.barplot(x='cut', y='price', hue='color', data=diamonds_df)
ax.legend(loc='upper right', ncol=4)
plt.show()
```

```python
#Modifying the font-size and rotation of x-axis
plt.figure(figsize = (10, 5))
ax = sns.barplot(x='cut', y='price', hue='color', data=diamonds_df)
ax.legend(loc='upper right',ncol=4)
# set fontsize and rotation of x-axis tick labels
ax.set_xticklabels(ax.get_xticklabels(), fontsize=13, rotation=30)

plt.show()
```

## Different Aspect of Bar Graph( Horizontal?) 

```python
plt.figure(figsize = (10, 5))
ax = sns.barplot(x="price", y="cut", hue='color', data=diamonds_df, estimator=np.median, orient = "h")#horizontal bar plot requres x= numerical data
plt.show()
```

## Anotation  on Plot

```python
ideal_df = diamonds_df.loc[diamonds_df['cut']=='Ideal']
x = ideal_df.index.tolist()[0]
y = len(ideal_df)
print(x)
print(y)
```

```python
plt.figure(figsize = (6,5))
ax = sns.catplot("cut", data=diamonds_df, aspect=1.5, kind="count", color="b")
plt.annotate(text = 'Say something for this location', xy=(x,y), xytext = (x+0.3, y+2000), arrowprops=dict(facecolor='red'))
plt.show()
```

## Creating a Scatter Plot

```python
# load a seaborn dataset
mpg_df = sns.load_dataset("mpg")
```

```python
mpg_df.head()
```

```python
ax = sns.scatterplot(x="weight", y="mpg", data = mpg_df)
```

```python
ax = sns.scatterplot(x="model_year", y="mpg", data=mpg_df, color = "darkgreen")
```

## Creating a Hexagonal Binning Plot

```python
sns.set(style="ticks")
ax = sns.jointplot(mpg_df.weight, mpg_df.mpg, kind="hex", color="#4CB391")
```

# Creating Contour Plot

```python
# contour plot
# generate KDE plot: first two parameters are arrays of X and Y coordinates of data points
# parameter shade is set to True so that the contours are filled with a color gradient based on number of data points
ax = sns.kdeplot(mpg_df.weight, mpg_df.mpg, shade=True)
```

# Creating a line plot

```python

ax = sns.lineplot(x="model_year", y="mpg", data = mpg_df)
```

```python
ax = sns.lineplot(x="model_year", y="mpg", data= mpg_df, ci=90) #ci= confidence interval or standard deviation
```

# Present Data Across Time With a Line Plot

```python
flights_df = sns.load_dataset("flights")
print(flights_df.head(12))
```

```python
plt.figure(figsize = (12,6))

ax = sns.lineplot(x="year", y="passengers", data=flights_df[flights_df['month']=='Jan'], color='green')
ax = sns.lineplot(x="year", y="passengers", data=flights_df[flights_df['month']=='Feb'], color='red')
ax = sns.lineplot(x="year", y="passengers", data=flights_df[flights_df['month']=='Mar'], color='blue')
ax = sns.lineplot(x="year", y="passengers", data=flights_df[flights_df['month']=='Apr'], color='cyan')
ax = sns.lineplot(x="year", y="passengers", data=flights_df[flights_df['month']=='May'], color='pink')
ax = sns.lineplot(x="year", y="passengers", data=flights_df[flights_df['month']=='Jun'], color='black')
ax = sns.lineplot(x="year", y="passengers", data=flights_df[flights_df['month']=='Jul'], color='grey')
ax = sns.lineplot(x="year", y="passengers", data=flights_df[flights_df['month']=='Aug'], color='yellow')
ax = sns.lineplot(x="year", y="passengers", data=flights_df[flights_df['month']=='Sep'], color='turquoise')
ax = sns.lineplot(x="year", y="passengers", data=flights_df[flights_df['month']=='Oct'], color='orange')
ax = sns.lineplot(x="year", y="passengers", data=flights_df[flights_df['month']=='Nov'], color='darkgreen')
ax = sns.lineplot(x="year", y="passengers", data=flights_df[flights_df['month']=='Dec'], color='darkred')
plt.title("Number of passengers in different months over different years")
plt.show()
```

## Creating and Exploring a Heatmap

**Important**: construction of pivot table

```python

df_pivoted = flights_df.pivot(index = "month", columns = "year", values = "passengers")
df_pivoted.head()
```

```python
ax = sns.heatmap(df_pivoted)
```

```python
plt.figure(figsize = (13,8))
ax = sns.heatmap(df_pivoted, annot = True,  annot_kws={"size": 7})
plt.show()
```

```python
ax = sns.clustermap(df_pivoted, col_cluster= False, row_cluster = True)
```
