## **Machine Learning Toolboxes**

- **[Basic Python](https://www.python.org/):** Provides basic python prgramming environment, data structures and operations. 

- **Python Extensions:**
  * **[NumPy](https://numpy.org/doc/stable/user/quickstart.html)**: Helps for numerical computations using n dimensional arrays.
  * **[Pandas](https://pandas.pydata.org/docs/user_guide/10min.html)**: Data frame package which helps for loading and handling data.
  * **[Matplotlib](https://matplotlib.org/stable/tutorials/index.html)**: Helps for plotting/visualization.
  * **[Seaborn](https://seaborn.pydata.org/introduction.html)**: Additional plotting/visualization package.
  * **[Scikit-Learn](https://scikit-learn.org/stable/user_guide.html)**: Scikit-learn library has several machine learning tools, including data preprocessing.
 
  * **[Tensorflow](https://www.tensorflow.org/)**: Tensorflow privides large scale compuation and visualization platform.
  
  * **[Keras API](https://keras.io/guides/)**: High level API for building Deep Learning models.
  
 
- **Environment and Platforms:**

   - **[Anaconda](https://www.anaconda.com/)**: This is a python package management system which install and mangages various python packages in a single platform.It is an important platform to run python locally (in your laptop). 
  
  - **[Jupyter](https://jupyter.org/)**: This is a Web Development Environment which helps to integrate python code, live result executation, visualization,  interpretation. Jupyter comes with anaconda installation.

  - **[LaTeX](https://www.overleaf.com/learn/latex/Learn_LaTeX_in_30_minutes)** Helps to write publication ready documentation, report, papers, thesis, etc
  - **[Google Drive](https://www.google.com/drive/)**: This helps to store class materials, quizzes, projects, data, etc.

  - **[Google Colab](https://colab.research.google.com/notebooks/intro.ipynb)**: This is a cloud computing platform offered by google, which helps to run python programming and data science projects without installing any python packages  in the local machines. Just we need a good internet connection. No matter what is the computing capacity of your laptop, you can have remote access to High Performance Computing Data Centers of Google powered by CPU, GPU, and even TPU.
